## Recurrent Neural Network Model

We could use a standard neural network (MLP) with few hidden layers.

Input all words and try to predict the names in the sequence.

Problems

- Input and outputs can be different lenghts in different examples. MLP network is fixed in size. Maybe we could pad but still not very ideal.

- Wouldn't share features learned across different positions. Since weights for timestamp 1 and 2 are completely disjoint for the input neuron. Similar to images where we want to learn a visual feature across different input positions.



What is an RNN ?? 


- input only $x_1$ and output $y_1$. 
- Then for step 2 $x_2$ output is calculated based on both $x_2$ and $a_1$ which is a hidden representation for $x_1$ (activation of $x_1$). 
- You repeat this for your whole input sequence.

He doesn't like the representations that are not unrolled. Like the rolled single block representation is a bit confusing.


- Vertical (y) and horizontal (a) connections are governed with separate weight matrices.

Weakness:

- Information from only up to $(t-1)$ timestamp is used (so preceeding words). This is a big limitation because we may need to know about subsequent words to make a prediction. BiRNN is a way to solve this limitation left to right and concatenated with right to left.


Example:

"He said, 'Teddy Roosevelt was a great President.'"


"He said, 'Teddy bears are on sale!'"


You need some more context to make prediction whether Teddy is referring to a human or not!!!




### Forward Propagation

$$
a^{<1>} = g(W_{aa} a^{<0>}+W_{ax}x^{<1>} + b_a)\\
\hat{y}^{<1>} = g(W_{ya} a^{<1>} + b_y)
$$

second index in $W_{ya}$ means second index is multiplied by $a$ type quantity.

For $a$ Activations used are commonly tanh or ReLU for `g`. For $y$ it depends on the task.


$$
a^{<t>}  = g(W_{aa} a^{<t-1>} + W_{ax} x^{<t>} + b_a)\\\
...
$$

Notice that $b_a$ and $b_y$ are same for each timestamp since it is the same block of weights applied over and over again on different x values.




#### Simplified RNN Notation

$W_{ax}$ and $W_{aa}$ are a bit redundant. So we will simply the notation.

$$
a^{<t>} = g(W_a[a^{<t-1>},x^{<t-1>}]+ b_a)\\
W_a = \begin{bmatrix}W_{aa} | W_{ax}\end{bmatrix}
$$

where $[a^{<t-1>},x^{<t-1>}]$ denotes vertical stacking of two vectors (not horizontal) so that we can still do the same matrix with vector multiplication.

Similarly lets just call $W_{ya}$ as $W_y$ since we know it will always be multiplied by a.


## Backpropagation Through Time

Programming frameworks will have you covered but still good to know..

So the main idea is we use all inputs up to $t-1$ to calculate $t$ output. 


Here Andrew discussed simple cross entropy loss and the fact that backward step goes in the opposite steps of forward. So, the most significant computational step is the backward for $a$ since we go $t$ steps backward. This is also called "Backward propagation through time".

So the basic task of language model is to output a likelihood for a given output sequence. Sequence of outputs not inputs



#### Language Modelling with an RNN

Training set: Large corpus of english text.

- Pad with \<EOS\>
- Tokenize each sentence
- Each token is a one-hot-encoding
- For out of vocab tokens we can use the same \<UNK\> token.
- So we will try to predict the next token given the previous context!!

Cats average 15 hours of sleep a day. \<EOS\>

**RNN Language Model.**

$y^{<1>}$ is Cats 

Predict $y^{<2>}$ where $x^{<2>}$ is $y^{<1>}$ and we also use the activation from previous step $a^{<1>}$. 

$$
P(\text{average}|\text{cats}) = ....
$$

repeat for each step.




Cost Function.

$$
\mathcal{L} = \sum_t ( - \sum_i y_i^{<t>} log(\hat{y}_i^{<i>}) )
$$

For calculating the likelihood of a sentence, simply multiply $P(y_1) \cdot P(y_2|y_1) \cdot P(y_t|y_1,y_2,y_{t-1})$. Probably we will use log likehood not to get to 0 immediately by multiplying everything.

## Sampling Novel Sequences

It is slightly different than training.



- Your first timestamp you randomly sample from softmax.
- Then given what you predicted, you use it as input to next timestamp and sample in the same way. So whatever you sampled becomes input to the second position.


That's why such approaches are called "Auto-regressive" model.

- You either stop until it generates the stop token or simply have a hard cutoff lets say at 20 tokens.

- Important note about \<UNK\>: Sometimes the above algorithm would output the unknown token. Simply handle this by rejecting and sample again. 



**Character-level Language Model**

- No worries about out ot vocabulary issues.

- Disadvantage: Sequence gets very long, character representation is kind of less meaningful compared to full words or tokens.

## Vanishing Gradients with RNNs


Lets have an Language modeling example with an RNN.



    The cat, which already ate ....... , was full 

                          vs 

    The cat(s), which already ate ....., (were) full



This is a classical example to illustrate the long-range dependencies in natural language. 


Recall that for very deep neural network we had vanishing gradient issue because we backpropagate so many times. RNNs have same problem as the sequence gets long!!


It is difficult to associate the loss from a far timestamp (such as mistake of was vs were).

So RNNs are heavily influenced from the local information. Such as couple words before it. This is the basic RNN issue. We will try to find solutions using other RNN architectures.


Apparently Exploding gradients are easier to detect because the numbers will simply get very large or even nan.

Solutions to exploding gradients:

- Gradient Clipping: If it is bigger than a certain threshold rescale your gradients

Vanishing grad has no such trivial solutions

## GRUs 

Effective solution to vanishing gradients. Makes it much better with long term dependencies.


**Formulation (Simplified)**

We will have a new concept of a memory cell (shown as $c$) that is carried our. You can think of it as the same thing with $a$ for now. So:

$$c^{<t>} = a^{<t>}$$

And remember that if you have 100 hidden units in your RNN, each of $c$ and $a$ are 100-dimensional vectors.


Intuition: 

- Previously at each step, we would have a matrix vector multiplication to get the new hidden activation $a^{<i>}$ and pass it to the $(i+1)^{th}$ step

- With GRUs, we don't directly pass the activation. Instead we use a gate to pass a weighted combination of the activation (or cell memory $c$) from previous timestamp to the $(i+1)^{th}$ computation.


Here is the mathematical way to formulate it:

(for this example we can just assume a=c)

$$
\hat{c}^{<t>} = g(W_c [c^{<t-1>}| x^{<t>}] + b_c)\\
\Gamma_u = \sigma(W_u [c^{<t-1>} | x^{<t>}] + b_u)\\
c^{<t>} = \Gamma_u * c^{<t>}  + (1-\Gamma_u) * c^{<t-1>}
$$

$\hat{c}^{<t>}$ is referred as the "candidate" memory cell. Later we use the gate function to decide whether or not OR how much we use this candidate.

Note that:

- We used sigmoid ($\sigma$) activation for the Gate (shown as Gamma $\Gamma$) so that it is between (0,1). So when we do weighted combination, it adds up to 1.
- $\Gamma_u$ (u denotes update) is also an 100 dimensional vector. So we turn off or turn on each bit from the previous memory separately. So the gamma value for each dimension is separate.
- If the Network thinks the memory should be kept the same it will generate $\Gamma_u=0$ for those cells and they would be passed on the next computation without change.

- Maybe some cell retain information about singularity/plurality and it will be passed without change. Maybe some other cells have information about food so they will be overriden to incorporate the latest context.



The above is the simplified GRU formulation that has the core "gated" idea.
However, in practice there is an alternative with "Relevance Gate" that researchers found out to perform better.

**Full GRU Formulation**

It is almost the same but when calculating the candidate vector $\hat{c}^{<t>}$, we utilize another weight matrix (Relevance gate) to decide how relevant is $c^{<t-1>}$ for computing the candidate vector.

$$
\Gamma_r = \sigma(W_r [c^{<t-1>}, x^{<t>}]+ b_r)\\
\hat{c}^{<t>} = tanh(W_c [\Gamma_r * c^<t-1> , x^{<t>}] + b_c)
...
$$

rest is identical... Note that $\Gamma_r * c^<t-1> $ is again element-wise multiplication that decides which bits are relevant given the new context!!


#### How does this help with Vanishing Gradients??

- It is very easy to set $\Gamma_u=0$ since sigmoid converge to 0 quickly in negative values. 
- This means that the gradient from later time can be easily passed down to $c^{<t-1>}$ since it is multiplied by a value very close to 1 (identity function).

- This allows the gradient to go back multiple steps without getting shrinked and vanish.

## LSTM 

Three core ideas mainly on top of GRU. 

- In addition to the update ($\Gamma_u$) gate, we use a forget Gate ($\Gamma_f$) to determine how much of $c^{<t-1>}$ we pass to new context (instead of just using $(1-\Gamma_u)$).

- Separate the idea of $a^{<t>}$ and $c^{<t>}$ so maintiain two activations passed down through time. Instead of using $c^{<t-1>}$, use $a^{<t-1>}$ to calculate the values for each Gate (see the formulas below).

- For the above point, use separate gate called output gate $\Gamma_o$ to generate $a^{<t>}$ using $c^{<t>}$

$$
\hat{c}^{<t>} = tanh(W_c [c^{<t-1>},x^{<t>}]+ b_c)\\
\Gamma_u = \sigma(W_u [a^{<t-1>},x^{<t>}]+ b_u)\\
\Gamma_f = \sigma(W_f [a^{<t-1>},x^{<t>}]+ b_f)\\
\Gamma_o = \sigma(W_o [a^{<t-1>},x^{<t>}]+ b_o)\\
c^{<t>} = \Gamma_u * \hat{c}^{<t>} + \Gamma_f * c^{<t-1>}\\
a^{<t>} = \Gamma_o * tanh(c^{<t>})
$$

## BiRNN

Bidirectional RNN: Get information from the future!!


Motivating example is the Teddy example.


    "He said, 'Teddy Roosevelt was a great President.'"

                               vs 
                               
    "He said, 'Teddy bears are on sale!'"


Core idea:

- Left to right context might not be enough to predict something. So we also generate right to left context. 

- For each input we will have two representations $\overrightarrow{a^{<t>}}$ and $\overleftarrow{a^{<t>}}$ and we use both when we make the prediction whether Teddy is a named entity or not.

- Notice that the left to right context and right to left index are computed separately so there is no cyclical graph in our computation graph.


In those old days BiRNN are usually used where each representation is calculated using LSTM. So they were called BiLSTM I remember!! Wow good old days


Reminder that this puts a limitation to real-life applications. If we want to have a real-time speech recognition model, we have to wait until the speaker stops talking so that we can generate the right to left representations.



## Deep RNNs

Probably just stack all these RNNs vertically to get complex representations.


- So remember now that we have a time axis (horizontal) that we generate representations for.

- On top of that we will also stack vertical layers and repeat the same thing multiple times.



**Notation**

$a^{[2]<3>}$ denotes activation of third timestamp in the second layer.


For layers larger than 1, the input $x^{<t>}$ becomes the output of the previous layer:

$$
a^{[2]<3>} = g(W_a^{[2]}[a^{[2]<2>},a^{[1]<3>}] + b_a^{2})
$$

So since there is giant backward dependency, we can not really stack 100 layers of RNN because the backward step will be super slooooow. (layer 100 depends on all 99 layers' all activations in the BiRNN case)


RNNs are computationally expensive!! So we can NOT do very deep RNNs
